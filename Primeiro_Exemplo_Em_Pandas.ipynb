```python
# Criar um dataframe de gas_df contendo 
# Apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
# Somente com combustivel sendo GASOLINA e exibir na tela

gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display (gas_df)
```

```python
# Exibir qual o posto mais caro 

display (gas_df[['Revenda','Municipio', 'Valor de Venda']].max())
```

```python
#DataFrame.loc[] com múltiplas condições para filtragem
#Quais são os preços, postos que vendem ETANOL na cidade (INDAIATUBA) 
#ordenado do menor valor de venda para o maior

etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
etanol_indaiatuba_df.sort_values(by='Valor de Venda', inplace=True)
display(etanol_indaiatuba_df)
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') & ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), ['Valor de Venda']].mean())
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO
# Usando ISIN

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') & (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), ['Valor de Venda']].mean())
```

```python
# Como mostrar media de valor de vendaa por COMBUSTIVEL no Brasil todo

med_por_comb_df = ca_df[['Produto','Valor de Venda']].groupby(by='Produto').mean().round(2)
display(med_por_comb_df)
```

```python
# Adicionar uma coluna de valor booleano no combustiveis_df chamada "Ativo"
# Que por padrão vai ser true para TODAS as linhas

combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())
```

```python
# Coluna nova chamada "Obs" que preencha automaticamente "MELHOR CIDAD
#quando o Municipio for igual SP, ARAPIRACA etc... 

#combustiveis_df['Obs'] = ["Melhor Cidade" if municipio == 'SAO PAULO' else '' for municipio in combustiveis_df['Municipio']]
#display (combustiveis_df['Municipio', 'Obs'].loc[combustiveis_df ['Municipio'].isin(['SAO PAULO']), ['Municipio', 'Obs']])


combustiveis_df['Obs'] = ["MELHOR CIDADE" if municipio == 'SAO PAULO' else None for municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO','INDAIATUBA', 'CAMPINAS', 'SALTO']), ['Municipio', 'Obs']])
```

```python
# (por Leandro Rodrigues)
# Se o valor de venda for maior que 6,5 reais, ele fala que tá Caro...caso contrário, estará barato

import numpy as np

combustiveis_df['Status do Valor de Venda']=np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Caro', 'Barato')
display (combustiveis_df[['Revenda', 'Valor de Venda', 'Status do Valor de Venda']])
```

```python
# Calcular postos de gasolina por habitantes temos na amostragem de combustiveis nov/2021
# documento = habitantes_estimado_ibge.csv

num_habitantes_df = pd.read_csv("habitantes_estimado_ibge.csv", sep=";")
num_habitantes_df.rename(columns={"Estado" : "Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
# Faz um MERGE dos dois dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitantes_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
# Destruir coluna completamente vazia(todas as linhas são nulas)

merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo', 'Status do Valor de Venda']

merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
display(merge_df.head(100))
```

```python
# Remover a linhas duplicadas

merge_df.drop_duplicates(inplace=True)
display (merge_df.head(100))
```

```python
# Agrupar e contar quantos postos tem na cidade 

postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "Números de Postos"}, inplace=True)
display(postos_por_municipio_df)
```

```python
#Agrupar e contar quantos postos tem na cidade

postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={'Revenda': 'Numero de Postos'}, inplace=True)
postos_por_municipio_df.reset_index(inplace=True)
display(postos_por_municipio_df)
```

```python
#Agrupar e contar quantos postos tem na cidade..
postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.reset_index(inplace=True)
#display(postos_por_municipio_df.info())
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "NumPostos"}, inplace=True)

postos_por_municipio_df['NumHabitantesPorPosto'] = postos_por_municipio_df['NumHabitantes2021'] / postos_por_municipio_df['NumPostos']
#display(postos_por_municipio_df.info())
display(postos_por_municipio_df)
```

```python
# Graficos
import matplotlib.pyplot as plt

plt.hist(combustiveis_df['Valor de Venda'])

#Vamos colocar um título no gráfico

plt.title("Preço dos combustíveis - Nov/2021")

#Rótulo horizontal e vertical

plt.xlabel("Preço (em reais)")
plt.ylabel("Quantidade de Coletas")

#Traça a linha vermelha tracejada com o preço médio
plt.axvline(combustiveis_df['Valor de Venda'].mean(), color='red', linestyle='dashed', linewidth=5)


#"Plota" o gráfico
plt.show()
```

```python
# Segundo
c_mean = combustiveis_df['Valor de Venda'].groupby(by=combustiveis_df['Produto']).mean()
display(c_mean)
```

```python
import seaborn as sns

#Definir a área do gráfico
plt.figure(figsize=(7,5))

#Plotar o gráfico
c_mean.plot(
    kind="barh",
    xlabel="Tipo de Combustível",
    ylabel="Preço reais/litro",
    title="Média de preços por combustível",
    color="red",
    alpha=0.3
)

#Grid
plt.grid()

#Remover as linhas da borda
sns.despine()

#Exibir
plt.show()
```

```python
from pandas.core.array_algos.replace import Pattern
 # Personalizar a planilha Excel

from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Color, Alignment

# Abrir o Excel no openpyxl
wb= load_workbook(excel) #wb= Workbook

#Pegar a planilha certa, Usando o Sheet Name (nome da planilha)

ws= wb ['Sumário'] # Work Sheet > planilha atual, ativa, de trabalho 

# Pintar o cabeçalho 

cinza = PatternFill("solid", fgColor="AAAAAA")
coords = ['A1', 'B1']
for coord in coords:
 ws[coord].fill = cinza

#Onde o preço do combustivel for maior ou igual a 6,5 reais (6.5) pinta a fonte de vermelho e deixa negrito

MAX_ROW = ws.max_row
num_linha = 2
while (num_linha <= MAX_ROW):
  coord = 'B'+str(num_linha) #coord="B{0}".format(num_linha)
  if ws[coord].value >= 6.5:
    ws[coord].font = Font(bold=True, color="FF0000", size= 12)
  else:
    ws[coord].font = Font(color='00FF00', size = 11)
  num_linha = num_linha + 1

# Salvar o Excel

wb.save(excel)
```
